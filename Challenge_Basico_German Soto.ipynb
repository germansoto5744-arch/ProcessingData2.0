##### Retiramos los valores nulos 

```python
mssng_percent = (terror.isna().sum() / len(terror)) * 100
mssng_percent = mssng_percent.sort_values(ascending=False)
mssng_percent.head(100)
```

```python
# Retiramos los valores nulos y convertimos la lista y creamos un nuevo dataframe. / Generamos una lista y date frame con columnas utiles
indice = 90
colukp = mssng_percent[mssng_percent < indice].index
colukp_list = colukp.to_list()
terror_limpio = terror[colukp_list]

print(f"Columnas: {len(terror.columns)}")
print(f"Columnas limpias: {len(terror_limpio.columns)}")
```

```python
terror_limpio.info()
```

##### Convertimos columna Kill y Wound en numerico

```python
terror_limpio['nkill'] = terror_limpio['nkill'].astype(float)
terror_limpio['nwound'] = terror_limpio['nwound'].astype(float)
```

### Realizamos el analisis descriptivo estadistico

```python
terror_limpio.describe()
```

#### Correlacion de los datos

```python
terror_no_nulls = terror_limpio[['nkill', 'nwound']].dropna()
correlation_matrix = terror_no_nulls.corr()
print(correlation_matrix)
```

##### Distribucion por Region

```python
fig = terror['region_txt'].value_counts().plot(kind='bar')
fig.update_layout(
    title="Distribucin de Incidentes por Regin",
    xaxis_title="Regin",
    yaxis_title="Nmero de Incidentes"
)
fig.show()
```

##### Distribucion por tipo de ataque

```python
fig = terror['attacktype1_txt'].value_counts().plot(kind='bar')

fig.update_layout(
    title="Distribucin de Tipos de Ataques",
    xaxis_title="Tipo de Ataque",
    yaxis_title="Nmero de Incidentes",
    width=1200,
    height=600
)
fig.show()
```

##### Top 10 pases por nmero de incidentes


```python
import plotly.express as px
data = terror['country_txt'].value_counts().head(10).to_pandas() 
fig = px.bar(
    x=data.values,
    y=data.index,
    orientation='h',
    labels={'x': 'Nmero de Incidentes', 'y': 'Pas'},
    color_discrete_sequence=['royalblue'],
    text=data.values
)

fig.update_traces(textposition='outside')

fig.show()
```
